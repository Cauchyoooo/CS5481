```python
import numpy as np
import matplotlib.pyplot as plt

# 给定的 CPI 指数数据
cpi_data = [240.007, 245.120, 251.107, 255.657, 258.811, 270.970, 292.655, 304.702]

# 年份列表，假设从 2016 年开始
years = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

# 进行 2016 - 2019 年数据的多项式拟合（这里使用二次多项式拟合）
cpi_data_2016_2019 = cpi_data[:5]
years_2016_2019 = years[:5]
coefficients = np.polyfit(years_2016_2019, cpi_data_2016_2019, 2)
polynomial = np.poly1d(coefficients)

# 生成从 2016 到 2023 的拟合曲线数据点
extended_years = np.arange(2016, 2024)
extended_fitted_curve = polynomial(extended_years)

# 绘制 CPI 数据，2016 - 2019 为红色，2020 - 2023 为蓝色
plt.plot(years_2016_2019, cpi_data_2016_2019, marker='o', linestyle='-', color='#880808', label='CPI Data (2016 - 2019) GOP')
plt.plot(years[4:], cpi_data[4:], marker='o', linestyle='-', color='#0047AB', label='CPI Data (2020 - 2023) DEM')

# 绘制拟合曲线为红色
plt.plot(extended_years, extended_fitted_curve, linestyle='--', color='#880808', label='Fitted Curve (2016 - 2023) GOP')

plt.title('U.S. CPI Index Trend from 2016 to 2023 with Fitted Curve')
plt.xlabel('Year')
plt.ylabel('CPI Index')
plt.legend()
plt.grid(True)

plt.savefig('us_cpi_trend_with_color_coded.png')

plt.show()
```
